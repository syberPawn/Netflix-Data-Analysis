# 📊 Netflix Data Analysis Project

This project explores and analyzes a dataset of Netflix titles to uncover insights about content types, trends, genres, contributors, and more using Python, Pandas, and Seaborn in a Jupyter Notebook.

---

## 📁 Dataset

- **Source**: Netflix Titles dataset
- **Format**: CSV
- **Columns** include:
  - `title`, `type`, `director`, `cast`, `country`, `date_added`, `release_year`, `rating`, `duration`, `listed_in`, `description`

---

## 📌 Objectives

1. Understand the structure and contents of the dataset
2. Clean and preprocess the data
3. Perform univariate and bivariate analysis
4. Derive deeper insights and trends
5. Visualize findings through plots
6. Summarize key takeaways and conclusions

---

## 🔍 Project Phases

### 🟢 Phase 1: Understand the Data
- Load and explore the structure
- Identify missing values

### 🟡 Phase 2: Clean the Data
- Handle missing values
- Convert data types
- Create new useful columns (e.g., `year_added`, `duration_int`)

### 🔵 Phase 3: Univariate Analysis
- Frequency counts
- Bar plots and distribution plots

### 🟣 Phase 4: Bivariate Analysis
- Analyze relationships between pairs of columns (e.g., type vs rating)
- Boxplots, group comparisons

### 🟠 Phase 5: Deeper Insights
- Content added over time
- Most popular genres
- Top directors, actors, and countries

### 🟣 Phase 6: Visualization & Summary
- Summary dashboards and visualizations
- Export cleaned dataset
- Key conclusions and possible business implications

---

## 📈 Tools & Libraries Used

- **Python 3**
- **Jupyter Notebook**
- **Pandas**
- **NumPy**
- **Matplotlib**
- **Seaborn**

---

## ✅ Results & Insights

- Movies dominate the content type on Netflix.
- A spike in content addition around 2019–2020.
- Certain genres like Dramas, Comedies, and Documentaries are most common.
- USA is the most common production country.
- Some directors and actors are highly frequent in Netflix content.
